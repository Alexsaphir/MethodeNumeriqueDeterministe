```python
def J3(P):
    x = P[0]
    y = P[1]
    z = P[2]
    return np.abs(( x - y )*( y - z )*( x - z))

def genUnifD3(d):
    return np.random.uniform(-1,1, d)

def trace3(N):
    X = zeros((N, 3))
    for i in range(N):
        U = genUnifD3(3)
        
        if(i == 0):
            X[i, :] = U
        else:
            if(J3(U)<J3(X[i-1])):
                X[i, :] = U
            else:
                X[i, :] = X[i-1, :]
    return X
    
```

```python
def J(b,x,y):
    r=0
    for j in unique(b):
        w=[i for i in range(len(b)) if b[i] == j]
        if len(w)>0:
            nx = np.mean(x[w])
            ny = np.mean(y[w])
            
            r = r + np.sum(np.linalg.norm([x[w]-nx, y[w]-ny]))
    return r
```
